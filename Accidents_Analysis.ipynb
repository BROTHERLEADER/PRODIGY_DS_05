```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset

file_path = 'RTA Dataset.csv'

df = pd.read_csv(file_path)

# Display the first few rows of the dataframe

df.head()
```

```python
df.dropna(inplace=True)  # Dropping rows with missing values

# Verify if there are any missing values left
df.isnull().sum()
```

```python
# Summary statistics
summary_stats = df.describe()
print(summary_stats)

# Unique values in categorical columns
categorical_columns = df.select_dtypes(include=['object']).columns
for col in categorical_columns:
    print(f"{col}: {df[col].nunique()} unique values")
```

```python

# Distribution of accidents by road conditions
road_condition_counts = df['Road_surface_conditions'].value_counts()
print(road_condition_counts)
```

```python

# Distribution of accidents by weather conditions
weather_condition_counts = df['Weather_conditions'].value_counts()
print(weather_condition_counts)
```

```python

# Distribution of accidents by  day of week
time_of_day_counts = df['Day_of_week'].value_counts()
print(time_of_day_counts)
```

```python

# Plotting road conditions
plt.figure(figsize=(10, 6))
sns.countplot(y='Road_surface_conditions', data=df, order=df['Road_surface_conditions'].value_counts().index)
plt.title('Distribution of Accidents by Road Conditions')
plt.xlabel('Count')
plt.ylabel('Road Condition')
plt.show()
```

```python
# Plotting weather conditions
plt.figure(figsize=(10, 6))
sns.countplot(y='Weather_conditions', data=df, order=df['Weather_conditions'].value_counts().index)
plt.title('Distribution of Accidents by Weather Conditions')
plt.xlabel('Count')
plt.ylabel('Weather Condition')
plt.show()
```

```python
# Plotting  day of week
plt.figure(figsize=(10, 6))
sns.countplot(y='Day_of_week', data=df, order=df['Day_of_week'].value_counts().index)
plt.title('Distribution of Accidents by Time of Day')
plt.xlabel('Count')
plt.ylabel('Time of Day')
plt.show()
```
