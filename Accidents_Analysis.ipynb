```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# Load the dataset

```python
file_path = 'RTA Dataset.csv'

df = pd.read_csv(file_path)

# Display the first few rows of the dataframe

df.head()
```

# Data Preprocessing

```python
# Convert 'Time' column to datetime and extract hour
df['Time'] = pd.to_datetime(df['Time'], format='%H:%M:%S', errors='coerce').dt.hour

# Convert 'Day_of_week' to categorical type
df['Day_of_week'] = pd.Categorical(df['Day_of_week'], categories=[
    'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'], ordered=True)

# Drop rows with NaN values in 'Time' column
df.dropna(subset=['Time'], inplace=True)

# Display basic information about the dataset
df.info()
```

# Data Visualization

```python
# Plot number of accidents by day of the week
plt.figure(figsize=(12, 6))
sns.countplot(x='Day_of_week', hue='Day_of_week',data=df, palette='viridis')
plt.title('Number of Accidents by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Accidents')
plt.show()
```

```python
# Plot number of accidents by hour of the day
plt.figure(figsize=(12, 6))
sns.countplot(x='Time', hue='Time',data=df, palette='viridis')
plt.title('Number of Accidents by Hour of the Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Accidents')
plt.show()
```

```python
# Plot accident severity distribution
plt.figure(figsize=(10, 6))
sns.countplot(x='Accident_severity',hue='Accident_severity', data=df, palette='viridis')
plt.title('Accident Severity Distribution')
plt.xlabel('Accident Severity')
plt.ylabel('Count')
plt.show()
```

```python
severity_counts = df['Accident_severity'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(severity_counts, labels=severity_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('viridis', len(severity_counts)), wedgeprops=dict(width=0.3))
plt.title('Accident Severity Distribution')
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(14, 7))
severity_by_day = pd.crosstab(df['Day_of_week'], df['Accident_severity'])
severity_by_day.plot(kind='bar', stacked=True, colormap='viridis', ax=plt.gca())
plt.title('Accident Severity by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Accidents')
plt.legend(title='Accident Severity')
plt.show()
```

```python
# Plot accident severity by weather conditions
plt.figure(figsize=(14, 8))
sns.countplot(x='Weather_conditions', hue='Accident_severity', data=df, palette='viridis')
plt.title('Accident Severity by Weather Conditions')
plt.xlabel('Weather Conditions')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Accident Severity')
plt.show()
```

```python
weather_counts = df['Weather_conditions'].value_counts()
plt.figure(figsize=(10, 8))
plt.pie(weather_counts, labels=weather_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('viridis', len(weather_counts)))
plt.title('Accidents by Weather Conditions')
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.
plt.show()
```

```python
# Plot accident severity by road surface conditions
plt.figure(figsize=(14, 8))
sns.countplot(x='Road_surface_conditions', hue='Accident_severity', data=df, palette='viridis')
plt.title('Accident Severity by Road Surface Conditions')
plt.xlabel('Road Surface Conditions')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Accident Severity')
plt.show()
```

```python
# Plot accident hotspots by area
plt.figure(figsize=(14, 8))
sns.countplot(x='Area_accident_occured', hue='Accident_severity', data=df, palette='viridis')
plt.title('Accident Hotspots by Area')
plt.xlabel('Area Accident Occurred')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.legend(title='Accident Severity')
plt.show()
```

```python
# Plot accident severity by driving experience
plt.figure(figsize=(14, 8))
sns.countplot(x='Driving_experience', hue='Accident_severity', data=df, palette='viridis')
plt.title('Accident Severity by Driving Experience')
plt.xlabel('Driving Experience')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Accident Severity')
plt.show()
```

```python
# Plot accident severity by educational level
plt.figure(figsize=(14, 8))
sns.countplot(x='Educational_level', hue='Accident_severity', data=df, palette='viridis')
plt.title('Accident Severity by Educational Level')
plt.xlabel('Educational Level')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Accident Severity')
plt.show()
```
